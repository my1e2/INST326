```python
import tkinter as tk
from tkinter import ttk
import os

class App(tk.Tk):
    def __init__(self, title, size):
        # Main setup
        super().__init__()
        self.title(title)
        self.geometry(f'{size[0]}x{size[1]}')
        self.minsize(size[0], size[1])

        # Initialize components
        self.task_manager = TaskManager()  # Initialize the task manager
        self.task_manager.load_tasks()  # Load tasks from text file

        self.create_widgets()  # Create the application widgets

        # Run the application
        self.mainloop()

    def create_widgets(self):
        # menu frame (for potential future expansion)
        self.menu_frame = Menu(self)
        self.menu_frame.place(x = 0, y = 0, relwidth = 0.3, relheight = 1)

        # main frame for the to-do list
        self.main_frame = ttk.Frame(self)
        self.main_frame.place(relx = 0.3, y = 0, relwidth = 0.7, relheight = 1)

        # entry for new tasks
        self.entry = ttk.Entry(self.main_frame) # placing within main frame
        self.entry.pack(pady = 10)

        # button frame for add and delete buttons
        self.button_frame = ttk.Frame(self.main_frame) # placing within main frame
        self.button_frame.pack(pady = 10)

        self.add_button = ttk.Button(self.button_frame, text = "Add Task", command = self.add_task)
        self.add_button.pack(side = 'left', padx = 5)

        self.delete_button = ttk.Button(self.button_frame, text = "Delete Task", command = self.delete_task)
        self.delete_button.pack(side = 'left', padx = 5)

        # frame to hold the task checkboxes
        self.tasks_frame = ttk.Frame(self.main_frame) # placing within main frame
        self.tasks_frame.pack(pady = 10)
        self.update_task_list()  # Display loaded tasks

    def add_task(self):
        task = self.entry.get()
        if task:
            self.task_manager.add_task(task)  # Use the TaskManager to add the task
            self.update_task_list()
            self.entry.delete(0, tk.END) # delete word from entry box after task is added so you can add more tasks

    def delete_task(self): # click checkbox and then delete task after
        task_index = self.get_checked_task_index()
        if task_index is not None:
            self.task_manager.delete_task(task_index)  # Use TaskManager to delete the task
            self.update_task_list() # updating list of tasks 

    def get_checked_task_index(self): # retrieve the amount of tasks that were checked off
        for i in range(len(self.task_manager.check_tasks)):
            if self.task_manager.check_tasks[i].get():
                return i # return checks if boxes are checked
        return None # return nothing if no boxes are checked


    def update_task_list(self):
        # Clear the existing task frames
        for widget in self.tasks_frame.winfo_children():
            widget.destroy()

        # Create a checkbox for each task
        for i in range(len(self.task_manager.tasks)):
            task = self.task_manager.tasks[i]
            check_task = self.task_manager.check_tasks[i]
            checkbox = ttk.Checkbutton(self.tasks_frame, text = task, variable = check_task)
            checkbox.pack()  # Align checkboxes to the left


class TaskManager:
    def __init__(self):
        self.tasks = [] # Holding tasks 
        self.check_tasks = []  # To hold the check variable for each task

    def load_tasks(self): # loading tasks from last time program was used
        if os.path.exists("tasks.txt"):
            with open("tasks.txt", "r") as file:
                self.tasks = [line.strip() for line in file if line.strip()]
                self.check_tasks = [tk.BooleanVar(value = False) for tasks in self.tasks]

    def save_tasks(self): # saving tasks for next use
        with open("tasks.txt", "w") as file:
            for task in self.tasks:
                file.write(task + "\n")

    def add_task(self, task): # adding tasks through task manager
        self.tasks.append(task)
        self.check_tasks.append(tk.BooleanVar(value = False))
        self.save_tasks()  # Save tasks to text file

    def delete_task(self, index): # deleting tasks through task manager
        if 0 <= index < len(self.tasks):
            del self.tasks[index]
            del self.check_tasks[index]
            self.save_tasks()  # Save updated task list to text file

class Menu(ttk.Frame): # menu class for potential use
    def __init__(self, parent):
        super().__init__(parent)
        self.create_widgets()

    def create_widgets(self):
        return  # placeholder for menu items

# Execution code
App('Class-based To-Do List', (400, 300))
```
