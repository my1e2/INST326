#### 2. Simple Quiz Game
>  Create a quiz game that asks the user a series of questions and keeps track of their score.

```python
import random

def simple_quiz_game():
    possible_questions = { 
        "What is the largest mammal in the world?": ["Blue Whale", "Whale"],
        "Who painted the Mona Lisa?": ["Leonardo Da Vinci", "Da Vinci", "Leonardo"],
        "What galaxy are we located in?": ["Milky Way"],
        "Which geometric shape has four equal sides and four right angles?": ["Square"],
        "Who was the first U.S. President?": ["George Washington", "Washington"],
        "Which famous play features a character named Romeo?": ["Romeo and Juliet", "Romeo & Juliet"]
    }
    # Assign score to zero now for increments later on when you get correct answers
    score = 0
    print("This is a quiz game. You answer the questions and receive a score for your correct answers!\n")

    # Shuffle the questions so they appear in a random order
    questions_list = list(possible_questions.items())
    random.shuffle(questions_list)

    # Ask the user for input (answer) for each question in the question list 
    for question, correct_answers in questions_list:
        answer = input(question + " ")

        # check from the pool of correct answers each time to see if the user answered properly
        if any(answer.strip().lower() == correct_answer.lower() for correct_answer in correct_answers):
            print("You got the correct answer!\n")
            score += 1
        else:
            print(f"Wrong answer! The right answer was {correct_answers[0]}.\n")

         # Display the current score after each question
        print(f"Current score: {score} out of {len(questions_list)}.\n")

    # Final score printed
    print(f"Final score: {score} out of {len(possible_questions)} correct!")
    # Provide personal feedback based on the final score you get
    if score == len(possible_questions):
        print("Excellent! You got all the questions correct!")
    elif score > len(possible_questions) // 2:
        print("Great job! You did above average!")
    else:
        print("Nice try! Keep going!")

# Call the function to start the quiz
simple_quiz_game()


    
```
