#### 1. Basic Encapsulation
> Create a <font color=green>Person</font> class with a private attribute <font color=green>_age</font>. Implement getter and setter methods for <font color=green>_age</font> to control access.

```python
class Person:
    def __init__(self, age = 0):
        self._age = age
# getter
    def get_age(self):
        return self._age
# setter
    def set_age(self, age): 
        if age >= 0:
            self._age = age
        else:
            print("Invalid")
# method call
person1 = Person(18)
print(person1.get_age())

person2 = Person(23)
print(person2.get_age())
    
        
        
```

#### 2. Encapsulation with Properties
>  Refactor the previous <font color=green>Person</font> class to use Python's <font color=green>@property</font> decorator for getter and setter methods.

```python
class Person:
    def __init__(self, age = 0):
        self._age = age
# getter
    @property
    def age(self):
        return self._age
# setter
    @age.setter
    def age(self, age): 
        if age >= 0:
            self._age = age
        else:
            print("Invalid")
# method call
person1 = Person(18)
print(person1.age)

person2 = Person(23)
print(person2.age)
    
```

#### 3. Private Methods
>  Create a class <font color=green>BankAccount</font> with private attributes <font color=green>_balance</font> and <font color=green>_transaction_fee</font>. Implement a private method <font color=green>_apply_fee()</font> that deducts the fee from the balance and a public method <font color=green>withdraw(amount)</font> that uses <font color=green>_apply_fee()</font> if the withdrawal amount is valid.

```python
class BankAccount:
    def __init__(self, balance = 0, transaction_fee = 0):
        self._balance = balance
        self._transaction_fee = transaction_fee

    def _apply_fee(self): # subtract transaction fee for each withdraw from balance
        self._balance -= self._transaction_fee

    def withdraw(self, amount): # withdraws if amount is above 0 and greater than balance in bank account
         if amount > 0 and amount + self._transaction_fee <= self._balance:
            self._balance -= amount
            self._apply_fee()
            print(f"Withdraw: {amount}, New Balance: {self._balance}")
         else: 
             print("Invalid withdrawal")
    def get_balance(self): # getter for encapsulation before method call 
        return self._balance

# method call
bank_account = BankAccount(300, 5)
print("Balance:", bank_account.get_balance())

bank_account.withdraw(40)
```

#### 5. Restricting Access with Encapsulation
>  Create a <font color=green>Student</font> class with a private attribute <font color=green>_grade</font> and a method to update the grade if it is within a range of 0 to 100.

```python
class Student:
    def __init__(self, grade):
        self._grade = grade

    def update_grade(self, updated_grade): # updating grade in specific range
        if 0 <= updated_grade <= 100:
            self._grade = updated_grade
            print(f"Updated grade: {updated_grade}")
        else: 
            print("Invalid")
# getter of student grades
    def get_grade(self):
        return self._grade

student1 = Student(93)
print("Initial grade:", student1.get_grade())

student1.update_grade(97)
            
```

#### 6. Encapsulation with Input Validation
>  Create a <font color=green>Temperature</font> class with a private attribute <font color=green>_celsius</font>. Implement a property for <font color=green>celsius</font> that converts negative values to 0.

```python
class Temperature:
    def __init__(self, temp = 0): # setting up attributes for conversion to 0 initially
        self._celsius = temp
        if temp < 0:
            self._celsius = 0
        else:
            self._celsius = temp
    @property
    def celsius(self):
        return self._celsius

    @celsius.setter # setting values for conversion to 0 for encapsulation
    def celsius(self, value):
         if value < 0:
            self._celsius = 0
         else:
            self._celsius = value

temp1 = Temperature(-37)
print(temp1.celsius)

temp2 = Temperature(37)
print(temp2.celsius)
```

#### 7. Encapsulation with Multiple Private Attributes
>  Create a <font color=green>Car</font> class with private attributes <font color=green>_speed</font> and <font color=green>_fuel</font>. Provide methods to accelerate, decelerate, and check fuel level.

```python
class Car:
    def __init__(self, fuel, speed):
        self._speed = speed
        self._fuel = fuel 
# accelerating based on fuel and speed count
    def accelerate(self, accelerate):
        if self._fuel > 0:
            self._speed += accelerate
            self._fuel = accelerate - 3
            print(f"Accelerated {accelerate} mph")
        else:
            print("No fuel")
# decelerating based on same parameters
    def decelerate(self, decelerate):
            self._speed -= decelerate
            print(f"Decelerated {decelerate} mph")

    def check_fuel_level(self): # returning fuel leftover after process
        print(f"Fuel level: {self._fuel} gallons")
        
# method calls 
car1 = Car(15, 75)
car1.accelerate(15)
car1.decelerate(23)
car1.check_fuel_level()


```

#### 8. Encapsulation with Default Values
>  Create a <font color=green>User</font> class with private attributes <font color=green>_username</font> and <font color=green>_status</font> (default to "active"). Provide methods to deactivate and reactivate the users.

```python
class User:
    def __init__(self, username): # setting boundaries
        self._username = username
        self._status = "active"

    def deactivate(self): # deactivating user
        
        self._status = "inactive"
        print(f"'{self._username}' deactivated.")

    def reactivate(self): # activating user
        
        self._status = "active"
        print(f"'{self._username}' reactivated.")

    def get_status(self): # determing status based on last activation attempt 
        
        return self._status

    def get_username(self):
        
        return self._username

user1 = User("myles")
user1.deactivate()
user1.reactivate()
user1.get_status()

        
```

#### 9. Encapsulation with Protected Attributes
>  Create a <font color=green>Product</font> class with protected attribute <font color=green>_price</font>. Allow subclasses to modify <font color=green>_price</font>, but it should remain inaccessible directly outside the class hierarchy.

```python
class Product:
    def __init__ (self, price):
        self._price = price

    def get_price(self): # getter and setter method for product price
     
        return self._price

    def set_price(self, price):
      
        if price >= 0:
            self._price = price
        else:
            print("Invalid")
 
class ModifyPrice(Product): # lowering price class
    def __init__(self, price, new):
        super().__init__(price)
        self.new = new

    def new_pricing(self):
        
        if self._price > 100:
            new_price = self._price - self._price * 0.3
            self.set_price(new_price)
            print(f"New price: {self.get_price()}")
        else:
            print("Invalid")
# method calls
price1 = Product(300)

price2 = ModifyPrice(300, 20)

price2.new_pricing()


```

#### 10. Encapsulation with Custom Methods
>  Create an <font color=green>Employee</font> class with a private attribute <font color=green>_salary</font>. Implement a method to add a bonus without directly modifying <font color=green>_salary</font> from outside the class.

```python
class Employee:
    def __init__(self, salary): 
        self._salary = salary

    def bonus_salary(self, bonus): # adding salary bonus 
        self._salary += bonus
        print(f"New Salary: {self._salary}")

    def get_salary(self):  # getter
        return self._salary  

# method calls
employee1 = Employee(75000)
employee1.bonus_salary(5000)
```
