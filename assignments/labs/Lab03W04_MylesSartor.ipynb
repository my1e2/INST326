#### 8. Database: SQLite
>  Write a program that creates a SQLite database named <font color=green>example.db</font>, creates a table named <font color=green>users</font>, and inserts a new user with the name "Alice" and age 25.


```python
import sqlite3

# Creating/connecting the SQLite database 
conn = sqlite3.connect('example.db')


cursor = conn.cursor()

# Creating a table named users
cursor.execute('CREATE TABLE IF NOT EXISTS users (id INTEGER PRIMARY KEY, name TEXT, age INTEGER)')

# Inserting a new user with the name Alice and age 25
cursor.execute('INSERT INTO users (name, age) VALUES (?, ?)', ("Alice", 25))

# Commit the changes made to users table each time (updates)
conn.commit()

conn.close()




               



```

#### 9. Database: Querying SQLite
>  Write a program that reads all rows from the <font color=green>users</font> table in the <font color=green>example.db</font> database and prints them to the console.

```python
import sqlite3

# Creating/connecting the SQLite database 
conn = sqlite3.connect('example.db')


cursor = conn.cursor()

# Selecting all data in table
cursor.execute('SELECT * FROM users')

# Fetch
users = cursor.fetchall()

# Print the results, user id handles duplicates (primary key), so each value has a unique identifier 
for user in users:
    print(f"User ID: {user[0]}, Name: {user[1]}, Age: {user[2]}")


conn.close()


```
