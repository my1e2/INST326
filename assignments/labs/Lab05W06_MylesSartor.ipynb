#### 2. Label Widget
>  Create a Tkinter window with a label that displays "Welcome to Tkinter!".

#### 6. Combining Widgets
>  Create a Tkinter window with a label, an entry widget, and a button. When the button is clicked, change the label text to "Hello, [name]!" where [name] is the text entered in the entry widget.

```python
import tkinter as tk
from tkinter import ttk

def print_entry_label():
    name = entry.get()
    
    title_label.config(text = f"Hello, {name}!")
    

window = tk.Tk()
window.geometry('500x500')
window.title("Hello Tkinter")
title_label = ttk.Label(master = window, text = "Welcome to Tkinter!", font = 'Calibri 24 bold')
title_label.pack()
entry = ttk.Entry(master = window)
entry.pack()
button = ttk.Button(master = window, text = 'Button', command = print_entry_label)
button.pack()
window.mainloop()
```

#### 7. Checkbox Widget
>  Create a Tkinter window with two checkboxes labeled "Option 1" and "Option 2". When a checkbox is checked or unchecked, print the state of the checkboxes to the console. (refer to the second program from the lecture for syntax on creating checkboxes.)

```python
import tkinter as tk
from tkinter import ttk

# Printing the state of BOTH checkboxes to the console each time an option is checked or unchecked
def checkbox_state():
    option_1_state = option_1_var.get()
    option_2_state = option_2_var.get()
    if option_1_state: # if true
        print("Option 1 is checked")
    else: 
        print("Option 1 is unchecked")    
    if option_2_state: # if true
        print("Option 2 is checked")
    else:
        print("Option 2 is unchecked")
    

window = tk.Tk()
window.geometry('500x500')
window.title("Hello Tkinter")
option_1_var = tk.IntVar()
option_2_var = tk.IntVar()
checkbox_1 = ttk.Checkbutton(master = window, text = 'Option 1', variable = option_1_var, command = checkbox_state)
checkbox_1.pack()
checkbox_2 = ttk.Checkbutton(master = window, text = 'Option 2', variable = option_2_var, command = checkbox_state)
checkbox_2.pack()
window.mainloop()
```

#### 9. and 10. Note App (6pts)
>  This exercise will create a window for creating a note, like in a note-taking app. Create a Tkinter window with three labels, an entry box, a text box, and a button. You may also want to create two frames. The first label is a title label that says "Create your note". The second label says "note title: " and goes with the entry box in the first frame.  The third label says "note text: " and goes with the text widget in the second frame. The button contains the text "submit". When the button is pressed, it prints the note title, note text, and current timestamp to the console.

```python
import tkinter as tk
from tkinter import ttk
from datetime import datetime
# note taking app functionality
def note(): # function to print note details 
    
    note_title = entry.get()
    note_text = text.get("1.0", tk.END)
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    
    print(f"Note Title: {note_title}")
    print(f"Note Text: {note_text}")
    print(f"Timestamp: {timestamp}\n")

# Initialize the main window functionality 
window = tk.Tk()
window.geometry('500x500')
window.title("Note-Taking App")

# Title
title_label = ttk.Label(master = window, text = "Note", font = "Calibri 24 bold")
title_label.pack(pady = 10)


frame_1 = ttk.Frame(master = window)
frame_1.pack()

frame_2 = ttk.Frame(master = window)
frame_2.pack()

# Frame 1, storing entry to give note a title
note_label = ttk.Label(master = frame_1, text = "Note Title: ")
note_label.grid(padx = 5, pady = 5)
entry = ttk.Entry(frame_1, width = 15)
entry.grid(padx = 5, pady = 5)

# Frame 2, storing text box for the actual content of the note
text_label = ttk.Label(master = frame_2, text = "Note Text: ")
text_label.grid(padx = 5, pady = 5)
text = tk.Text(frame_2, height = 5, width = 15)
text.grid(padx = 5, pady = 5)

# Button to print note title, text, and current time to console
button = ttk.Button(master = window, text = "Note", command = note)
button.pack()


window.mainloop()

```
