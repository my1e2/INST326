#### 1. Exercise: Create a Vehicle Abstract Class
>  Define an abstract class <font color=green>Vehicle</font> with an abstract method <font color=green>move()</font>. Implement <font color=green>Car</font> and <font color=green>Bicycle</font> classes that override the <font color=green>move()</font> method
.


```python
from abc import ABC, abstractmethod

# abstract class 
class Vehicle(ABC):
    @abstractmethod
    def move(self):
        pass

# Car subclass 
class Car(Vehicle):
    def move(self):
        print("The car moves")

# Bicycle subclass 
class Bicycle(Vehicle):
    def move(self):
        print("The bicycle moves")

# method calls
car = Car()
bike = Bicycle()

car.move() 
bike.move()  

```

#### 2. Exercise: Create an Abstract Class for Appliances

```python
from abc import ABC, abstractmethod

# abstract class
class Appliance(ABC):
    @abstractmethod
    def turn_on(self):
        pass

# WashingMachine subclass 
class WashingMachine(Appliance):
    def turn_on(self):
        print("The washing machine turns on")

# Refrigerator subclass 
class Refrigerator(Appliance):
    def turn_on(self):
        print("The refrigerator turns on")

# method calls
washing_machine = WashingMachine()
refrigerator = Refrigerator()

washing_machine.turn_on()
refrigerator.turn_on()  

```

```python
from abc import ABC, abstractmethod

# abstract class
class Database(ABC):
    @abstractmethod
    def connect(self):
        pass

# MySQLDatabase subclass
class MySQLDatabase(Database):
    def connect(self):
        print("Connecting to MYSQL")

# SQLiteDatabase subclass 
class SQLiteDatabase(Database):
    def connect(self):
        print("Connecting to SQLite")

# method calls 
mysql_database = MySQLDatabase()
sqlite_database = SQLiteDatabase()

mysql_database.connect()  
sqlite_database.connect()  
```

#### 5. Exercise: Implement Abstract Class for Shipping
> Create an abstract class <font color=green>Shipping</font> with an abstract method <font color=green>calculate_cost()</font>. Implement <font color=green>AirShipping</font> and <font color=green>GroundShipping</font>.

```python
from abc import ABC, abstractmethod

# abstract class
class Shipping(ABC):
    @abstractmethod
    def calculate_cost(self, distance):
        pass

# AirShipping subclass
class AirShipping(Shipping):
    def calculate_cost(self, distance):
        cost = distance * 3  
        print(f"Air shipping cost: {cost} dollars")

# GroundShipping subclass 
class GroundShipping(Shipping):
    def calculate_cost(self, distance):
        cost = distance * 2  
        print(f"Ground shipping cost: {cost} dollars")

# method calls
air_shipping = AirShipping()
ground_shipping = GroundShipping()

air_shipping.calculate_cost(15) 
ground_shipping.calculate_cost(15)  

```

#### 8. Exercise: Abstract Class for Appliances with Power Consumption
> Create an abstract class <font color=green>Appliance</font> that has an <font color=green>__init__</font> method with attributes <font color=green>brand</font> and <font color=green>power</font>. The class should have an abstract method <font color=green>calculate_annual_power_usage()</font> that calculates the appliance's yearly power consumption. Implement two subclasses: <font color=green>WashingMachine</font> and <font color=green>Refrigerator</font>, each with its own power usage calculation logic.


```python
from abc import ABC, abstractmethod

# abstract class
class Appliance(ABC):
    def __init__(self, brand, power):
        self.brand = brand     
        self.power = power     

    @abstractmethod
    def calculate_annual_power_usage(self):
        pass

# WashingMachine subclass 
class WashingMachine(Appliance):
    def __init__(self, brand, power, cycles):
        super().__init__(brand, power)
        self.cycles = cycles

    def calculate_annual_power_usage(self):
       
        yearly_use = self.power * self.cycles * 52 # weeks a year (once a week wash throughout year)
        print(f"{self.brand} Washing Machine yearly power usage: {yearly_use} watts")

# Refrigerator subclass 
class Refrigerator(Appliance):
    def __init__(self, brand, power, hours = 24): # 24 hours a day 
        super().__init__(brand, power)
        self.hours = hours

    def calculate_annual_power_usage(self):
        
        annual_usage = self.power * self.hours * 365 # days in a year (runs everyday throughout year)
        print(f"{self.brand} Refrigerator yearly power usage: {annual_usage} watts")

# method calls
washing_machine = WashingMachine("Speed Queen", 250, cycles = 6)
refrigerator = Refrigerator("Insignia", 150)

washing_machine.calculate_annual_power_usage()
refrigerator.calculate_annual_power_usage()


```

#### 9. Exercise: Abstract Class for Vehicles with Fuel Efficiency
> Create an abstract class <font color=green>Vehicle</font> that has an <font color=green>__init__</font> method with attributes <font color=green>make</font>, <font color=green>model</font>, and <font color=green>fuel_efficiency</font> (in miles per gallon). Include an abstract method <font color=green>calculate_range()</font> that calculates the range of the vehicle based on fuel efficiency and fuel tank size. Implement two subclasses, <font color=green>Car</font> and <font color=green>Motorcycle</font>, each with its own fuel tank size.


```python
from abc import ABC, abstractmethod

# abstract class
class Vehicle(ABC):
    def __init__(self, make, model, fuel_efficiency):
        self.make = make                  
        self.model = model                
        self.fuel_efficiency = fuel_efficiency  # Fuel efficiency in miles per gallon 

    @abstractmethod
    def calculate_range(self):
        pass

# Car subclass 
class Car(Vehicle):
    def __init__(self, make, model, fuel_efficiency, tank_size = 15):
        super().__init__(make, model, fuel_efficiency)
        self.tank_size = tank_size       

    def calculate_range(self):
        # range based on fuel efficiency and tank size
        range = self.fuel_efficiency * self.tank_size
        print(f"{self.make} {self.model} car range: {range} miles")

# Motorcycle subclass 
class Motorcycle(Vehicle):
    def __init__(self, make, model, fuel_efficiency, tank_size = 3):
        super().__init__(make, model, fuel_efficiency)
        self.tank_size = tank_size        

    def calculate_range(self):
        # range based on fuel efficiency and tank size
        range = self.fuel_efficiency * self.tank_size
        print(f"{self.make} {self.model} motorcycle range: {range} miles")

# method calls
car = Car("Nissan", "Altima", fuel_efficiency = 45)
motorcycle = Motorcycle("Kawasaki", "Ninja", fuel_efficiency = 90)

car.calculate_range()
motorcycle.calculate_range()

```

#### 10. Exercise: Abstract Class for Employees with Monthly Salary
> Create an abstract class Employee with an <font color=green>__init__</font> method that includes <font color=green>name</font>, <font color=green>id_number</font>, and <font color=green>base_salary</font> attributes. Add an abstract method <font color=green>calculate_monthly_salary()</font> that calculates the total monthly salary based on any additional factors. Implement two subclasses, <font color=green>Manager</font> (with a bonus) and <font color=green>SalesPerson</font> (with commission on sales).


```python
from abc import ABC, abstractmethod

# abstract class
class Employee(ABC):
    def __init__(self, name, id_number, base_salary):
        self.name = name                 
        self.id_number = id_number       
        self.base_salary = base_salary   

    @abstractmethod
    def calculate_monthly_salary(self):
        pass

# Manager subclass 
class Manager(Employee):
    def __init__(self, name, id_number, base_salary, bonus): # bonus to salary
        super().__init__(name, id_number, base_salary)
        self.bonus = bonus               

    def calculate_monthly_salary(self):

        monthly_salary = self.base_salary + self.bonus
        print(f"{self.name} monthly salary: ${monthly_salary}")

# SalesPerson subclass
class SalesPerson(Employee):
    def __init__(self, name, id_number, base_salary, commission, sales): # commission rate on sales
        super().__init__(name, id_number, base_salary)
        self.commission = commission 
        self.sales = sales            

    def calculate_monthly_salary(self):
        
        commission = self.commission * self.sales
        monthly_salary = self.base_salary + commission
        print(f"{self.name} monthly salary: ${monthly_salary}")

# method calls 
manager = Manager("Myles", 1, base_salary = 75000, bonus = 1000)
sales_person = SalesPerson("Brandon", 2, base_salary = 50000, commission = 10, sales = 43)

manager.calculate_monthly_salary()
sales_person.calculate_monthly_salary()

```
