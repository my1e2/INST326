#### 10. Nested Loops
>  Write a nested loop that prints a 3x3 matrix of asterisks.

```python
symbol = '*'
# looping through each row
for i in range(3):
    row = ""
# looping through each column
    for j in range(3):
        row += symbol + " " # adding an asterisk each iteration, followed by a space to make the 3x3 matrix
    print(row) # printing the asterisk iteration each time (for range 3)
    
  
        
```
